```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

i)

```python
#### Two sample t test

# Generate two samples of data
np.random.seed(231)
sample_size1 = 100
sample_mean1 = 20
sample_std1 = 5
sample_data1 = np.random.normal(loc=sample_mean1, scale=sample_std1, size=sample_size1)

sample_size2 = 100
sample_mean2 = 25
sample_std2 = 6
sample_data2 = np.random.normal(loc=sample_mean2, scale=sample_std2, size=sample_size2)
```

iii) visualize distribustion of each group

```python
plt.figure(figsize=(8, 6))
plt.hist(sample_data1, bins=20, color='blue', alpha=0.5)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution of Group 1')
plt.show()
```

```python

# Plot histogram for Group 2
plt.figure(figsize=(8, 6))
plt.hist(sample_data2, bins=20, color='red', alpha=0.5)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution of Group 2')
plt.show()
```

iv) interpretition of my result

```python

# Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(sample_data1, sample_data2, equal_var=False)  # Assuming unequal variances

# Define significance level
alpha = 0.05

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis. There is enough evidence to conclude that the means of the two samples are significantly different.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the means of the two samples are significantly different.")
```
